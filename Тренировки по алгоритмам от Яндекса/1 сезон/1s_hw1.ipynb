```python
a = list(map(int, input().split()))
t_room = a[0]
t_cond = a[1]
t = 0

reg = input()

if reg == "freeze":
    if t_room >= t_cond:
        t = t_cond
    else:
        t = t_room 
    
elif reg == "heat":
    if t_room >= t_cond:
        t = t_room
    else:
        t = t_cond
elif reg == "auto":
    t = t_cond
elif reg == "fan":
    t = t_room
# else:
#     print("Нет такого режима")
       

print(t)
```

```python

def fink(mode, troom, tcond):
    if mode == "fan":
        print(troom)
    elif mode == "auto":
        print(tcond)
    elif mode == "heat":
        if troom <= tcond:
            print(tcond)
        else:
            print(troom)
    elif mode == "freeze":
        if troom > tcond:
            print(tcond)
        else:
            print(troom)


def main():
    troom, tcond = map(int, input().split())
    mode = input()
    fink(mode, troom, tcond)


if __name__ == "__main__":
    main()
```
