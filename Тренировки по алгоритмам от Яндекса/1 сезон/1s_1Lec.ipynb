```python
# решение 1 O(N^2) - проход по строке для каждого символа в строке
s = input()
ans = '' #создали пустую строку
anscnt = 0
for i in range(len(s)):
    nowcnt = 0
    for j in range(len(s)):
        if s[i] == s[j]:
            nowcnt += 1
    if nowcnt > anscnt:
        ans = s[i]
        anscnt = nowcnt
print(ans)
```

```python
# решение 2 O(Nk)- проход по строке для каждого символа в множестве
s = input()
ans = '' #создали пустую строку
anscnt = 0
for now in set(s):
    nowcnt = 0
    for j in range(len(s)):
        if now == s[j]:
            nowcnt += 1
    if nowcnt > anscnt:
        ans = now
        anscnt = nowcnt
print(ans)
```

```python
# решение 3 O(N+k)=O(N) - использование словаря
s = input()
ans = '' #создали пустую строку
anscnt = 0
dct = {}
for now in s:
    if now not in dct:
        dct[now] = 0
    dct[now] += 1
for key in dct:
    if dct[key] > anscnt:
        anscnt = dct[key]
        ans = key

print(ans)       

```

```python
# Сумма последовательности
seq = list(map(int, input().split())) #чтение последовательности символов
# Считываем последовательность строки и делится на символы, 
# преобразутся в число и из всего из этого список
seqsum = 0
for i in range(len(seq)):
    seqsum += seq[i]
print(seqsum)
```

```python
#Максимум последовательности
seq = list(map(int, input().split()))
if len(seq) == 0:
    print("Отрцательные числа")
else:
    seqmax = seq[0]
    for i in range(1, len(seq)):
        if seq[i] > seqmax:
            seqmax = seq[i]
print(seqmax)
```
